```python
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

```python
def hiperbolic_paraboloid():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection="3d")

    x = np.linspace(-10, 10, 100)
    y = np.linspace(-10, 10, 100)
    X, Y = np.meshgrid(x, y)
    Z = X**2 - Y**2

    surface = ax.plot_surface(X, Y, Z, cmap='plasma', edgecolor='none', alpha=0.7)
    ax.plot_wireframe(X, Y, Z, color='k', linewidth=0.5)
    ax.view_init(20, 125)
    ax.set_axis_off()
    plt.show()

hiperbolic_paraboloid()
```
